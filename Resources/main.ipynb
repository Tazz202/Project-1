```python
#dependencies
import pandas as pd

```

```python
#CSV file path
gdp_file = "Resources/gdp2020.csv"

food_file = "Resources/foodbank2020.csv"


gdp = pd.read_csv(gdp_file)

food = pd.read_csv(food_file)

gdp.head()

```

```python
gdp = gdp.rename(columns={"Country Name" : "Country"})
gdp.head()
```

```python
food.head()
```

```python
combined = pd.merge(food, gdp, how='left', on= 'Country')
combined.head()
```

```python
del combined["Unnamed: 0_x"]
combined.head()
```

```python
del combined["Unnamed: 0_y"]
combined.head()
```

```python
combined["GDP Per Capita"] = combined["2020 GDP in USD"] / combined["Population Size"]
combined.head()

```

```python
combined.to_csv("Resources/combined.csv")
```
