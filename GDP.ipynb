```python
#dependencies
import pandas as pd
```

```python
#CSV file path
gdp_file = "Resources/API_NY.GDP.MKTP.CD_DS2_en_csv_v2_4413598.csv"

#Read using pandas
raw_gdp = pd.read_csv(gdp_file)
raw_gdp.head()
```

```python
#start up clean up process by only narrowing down to year 2020
raw_gdp.columns
```

```python
clean_gdp = raw_gdp.loc[:, ["Country Name", "Indicator Name", "2020"]]
clean_gdp.head()
```

```python
len(clean_gdp["2020"])
```

```python
#getting rid of zero values
gdp = clean_gdp.loc[(clean_gdp["2020"] > 0)]
len(gdp["2020"])

```

```python
gdp.head()
```

```python
#checking for duplicate values
duplicate_gdp = gdp[gdp.duplicated()]
duplicate_gdp
```

```python
#renaming Indicator Name column
gdp = gdp.rename(columns={"Indicator Name" : "GDP Measured In"})
gdp.head()
```
